## パッケージのインストール

```python
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

from PIL import Image
import matplotlib.pyplot as plt

import torchvision.transforms as transforms
import torchvision.models as models

import copy

import os
```

## デバイスの選択

GPUを使う場合，様々な変数に".to(device)"と付けなければいけないが，これは『".to(device)"の付いた変数をGPUへ運ぶ』という意味がある．\
つまり，GPUで計算する際に必要そうな変数(入力データ，モデル，標準化など)には付けなければならない．

```python
# "GPU"か"CPU"か選択
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

## Image画像の読み込み

```python
content_filename = "\content2.jpg"
style_filename = "\style4.jpg"

imsize = 300 if torch.cuda.is_available() else 128

loader = transforms.Compose([ # 一連の処理を一気に行う箱
    transforms.Resize([imsize,imsize]), # 入力画像をリサイズする
    transforms.ToTensor()])    # Pytorch用にTensor型に変換する

def image_loader(image_name):
    image = Image.open(image_name) # 画像の読み込み
    image = loader(image).unsqueeze(0) # 1次元目に次元を増やす
    return image.to(device, torch.float) # GPU

content_img = image_loader(os.getcwd()+ "\picture" + content_filename)
style_img = image_loader(os.getcwd() + "\picture" + style_filename)

assert content_img.size() == style_img.size(), (
    "We need to import content and style images of the same size")

print(content_img.size())
print(style_img.size())
```

## 画像の表示

```python
# ↓これが何なのか...PILに変換するらしいけど。
unloader = transforms.ToPILImage()

plt.ion() # 対話モードでも画像を表示できる

def imshow(tensor, title=None):
    image = tensor.cpu().clone() # we clone the tensor not to do changes on it
    image = image.squeeze(0) # [1,3,256,256] -> [3,256,256]に変換
    image = unloader(image) # PIL型に...
    plt.imshow(image)
    if title is not None:
        plt.title(title)
    plt.pause(0.001) # プロットが更新されるよう小停止

plt.figure()
imshow(content_img, title="content Image")

plt.figure()
imshow(style_img, title="Style Image")
```

## 評価関数

### Content Loss

```python
class ContentLoss(nn.Module):
    def __init__(self, target):
        super(ContentLoss, self).__init__() # 親クラスのインスタンスメソッドの実行
        self.target = target.detach() # variable型 -> tensor型へ
        
    def forward(self, input):
        self.loss = F.mse_loss(input, self.target)
        return input # -> 多分、入力(生成画像)を更新していくためにreturnがinputになっている
    
```

### Style Loss

```python
class StyleLoss(nn.Module):
    def __init__(self, target_feature):
        super(StyleLoss, self).__init__()
        self.target = gram_matrix(target_feature).detach()
    
    def forward(self, input):
        G = gram_matrix(input)
        self.loss = F.mse_loss(G, self.target)
        return input

def gram_matrix(input):
    a, b, c, d = input.size()
    # a : batch size (=1)
    # b : number of feature maps
    # c : height of a f. map
    # d : width of a f. map
    #(c,d) = dimensions of a f. map (N=c*d)
    
    features = input.view(a*b, c*d)
    
    G = torch.mm(features, features.t()) # スタイル行列の計算
    
    return G.div(a*b*c*d) # 標準化。次元数が大きくなるほどGも大きくなるから。
```

## CNNの学習済みモデル(VGG16 or 19)のインポート

```python
# VGG19のインポート
cnn = models.vgg19(pretrained=True).features.to(device).eval()
```

## 入力画像の標準化

 VGG19は入力画像が，mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]で標準化された後に学習されたモデルなので，テストで使う際にも入力データを以上の条件で標準化する必要がある．

```python
cnn_normalization_mean = torch.tensor([0.485, 0.456, 0.406]).to(device)
cnn_normalization_std = torch.tensor([0.229, 0.224, 0.225]).to(device)

class Normalization(nn.Module):
    def __init__(self, mean, std):
        super(Normalization, self).__init__()
        self.mean = torch.tensor(mean).view(-1,1,1)
        self.std = torch.tensor(std).view(-1,1,1)
        
    def forward(self, img):
        return (img - self.mean) / self.std
```

## モデルと評価関数値の取得

```python
# 抜き出す層の指定
content_layers_default = ["conv_4"]
style_layers_default = ["conv_1","conv_2","conv3","conv_4","conv_5"]

def get_style_model_and_losses(cnn, normalization_mean, normalization_std,
                             style_image, content_img,
                             content_layers = content_layers_default,
                             style_layers = style_layers_default):
    cnn = copy.deepcopy(cnn) # <- 多分，外のモデルの値が変わったりしないようにするためかと．
    
    normalization = Normalization(normalization_mean,normalization_std).to(device)
    
    content_losses = []
    style_losses = []
    
    # VGG19から必要な層だけを取り出して，新たなCNNモデルを構築する
    # nn.Sequential()については，以下を参照↓
    # https://www.hellocybernetics.tech/entry/2017/10/20/025702#nnModule
    model = nn.Sequential(normalization) # <- 初めに標準化をいれ，後から層を追加していく
    
    i = 0 # 畳み込み層と遭遇する度，カウントされる
    for layer in cnn.children():
        if isinstance(layer, nn.Conv2d): # isinstance()は型判定
            i += 1
            name = "conv_{}".format(i) # formatメソッド．{}に()の値が入る
        elif isinstance(layer, nn.ReLU):
            name = "relu_{}".format(i)
            layer = nn.ReLU(inplace = False)
        elif isinstance(layer, nn.MaxPool2d):
            name = "pool_{}".format(i)
        elif isinstance(layer, nn.BatchNorm2d):
            name = "bn_{}".format(i)
        else:
            raise RuntimeError("Unrecognized layer:{}".format(layer.__class__.__name__))
        
        model.add_module(name, layer)
        
        # content_layers(リスト)の中にname(変数)がいるかどうか
        if name in content_layers:
            # content loss をモデルに加える
            target = model(content_img).detach()
            content_loss = ContentLoss(target)
            model.add_module("content_loss_{}".format(i), content_loss)
            content_losses.append(content_loss)
            
        if name in style_layers:
            # style loss をモデルに加える
            target_feature= model(style_img).detach()
            style_loss = StyleLoss(target_feature)
            model.add_module("style_loss_{}".format(i), style_loss)
            style_losses.append(style_loss)
     
    # 全結合層など使わない層もモデルに含まれているので，それらの層を削除する
    for i in range(len(model) - 1, -1, -1):
        if isinstance(model[i], ContentLoss) or isinstance(model[i], StyleLoss):
            break
    model = model[:(i + 1)]
    
    return model, style_losses, content_losses
```

## ホワイトノイズ(生成対象画像)の生成

```python
# ホワイトノイズの生成
input_img = torch.randn(content_img.data.size(), device=device)
# もし，学習を早めたいのであれば次のコードを使って
#input_img = content_img.clone()

plt.figure()
imshow(input_img, title='Input Image')
```

## 最適化の設定
評価関数値が下がるように入力データ(ホワイトノイズ)の値を更新していくので，optimには入力データ(ホワイトノイズ)を入力し，評価関数値を入力データ(ホワイトノイズ)で偏微分し，入力データ(ホワイトノイズ)の値を更新することで，画像生成を行う(ここでは，微分値の取得までしかしていない)．\
ここでは，評価関数値に対し，『LBFGSという方法とinput_imgを使って偏微分します』と宣言しただけ．

```python
def get_input_optimizer(input_img):
    optimizer = optim.LBFGS([input_img.requires_grad_()])
    return optimizer
```

## Neural Style Transfer の実装

```python
def run_style_transfer(cnn, normalization_mean, normalization_std,
                       content_img, style_img, input_img, num_steps=300,
                       style_weight=[5e5,5e5,5e5,5e5,5e5], content_weight=[1]):
    print("Building the style transfer model...")
    model, style_losses, content_losses = get_style_model_and_losses(cnn,
        normalization_mean, normalization_std, style_img, content_img)
    optimizer = get_input_optimizer(input_img)
    
    print("Optimizing...")
    run = [0]
    
    while run[0] <= num_steps:
        
        def closure():
            # input_imgが[0,1]を超える範囲で最適化されることがあるので，[0,1]に毎回修正する
            input_img.data.clamp_(0,1)
            
            optimizer.zero_grad()
            model(input_img)
            style_score = 0
            content_score = 0
            
            for i, sl in enumerate(style_losses):
                style_score += (sl.loss * style_weight[i])
            for j,cl in enumerate(content_losses):
                content_score += (cl.loss * content_weight[j])
            
            loss = style_score + content_score
            loss.backward() # 勾配の計算
            
            run[0] += 1
            if run[0] % 50 == 0:
                print("run {}:".format(run))
                print('Style Loss : {:4f} Content Loss: {:4f}'.format(
                    style_score.item(), content_score.item()))
                print()
            
            return style_score + content_score
        
        optimizer.step(closure)
    
    input_img.data.clamp_(0,1)
    
    return input_img
    
```

## Neural Style Transfer の実行

```python

output = run_style_transfer(cnn, cnn_normalization_mean, cnn_normalization_std,
                            content_img, style_img, input_img)

plt.figure()
imshow(output, title='Output Image')

# sphinx_gallery_thumbnail_number = 4
plt.ioff()
plt.show()
```
